```python
import numpy as np
import matplotlib.pyplot as plt
```

## Was ist ein Polardiagram?

```python
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.set_thetamin(0)
ax.set_thetamax(180)
ax.set_theta_zero_location('N')
ax.set_xticks(np.deg2rad(range(0, 181, 45)))

image = plt.imread('../figures/ship2.png')
axicon = fig.add_axes([.45,0.33,0.4,0.4], frameon=False)
axicon.imshow(image)
axicon.set_xticks([])
axicon.set_yticks([])

kw_annotate = {'xycoords': 'axes fraction',
               'textcoords': 'axes fraction', 
               'ha': 'center', 'va': 'center',
               'size': 15,
               'bbox': dict(boxstyle="round", fc=(1.0, 0.7, 0.7), ec=(1., .5, .5)),
               'arrowprops':dict(arrowstyle="fancy",
                                fc=".3", ec="none",
                                connectionstyle="angle3,angleA=0,angleB=-90"),
              }

ax.annotate('vor dem Wind', xy=(.75, .35), xytext=(0.75, .08), **kw_annotate)
ax.annotate('halber\nWind', xy=(.6, .5),xytext=(.35,.5), **kw_annotate)
ax.annotate('im Wind', xy=(.75, .65), xytext=(.75,.95), **kw_annotate)

ax.plot(np.deg2rad(170 + 10.*np.random.random(50)), np.random.normal(3., .5, 50),'y<')
ax.plot(np.deg2rad(150 + 20.*np.random.random(50)), np.random.normal(3., .5, 50),'b<')
ax.plot(np.deg2rad(130 + 20.*np.random.random(50)), np.random.normal(2.5, .5, 50),'r<')
ax.plot(np.deg2rad(110 + 20.*np.random.random(50)), np.random.normal(2., .5, 50),'g<')
ax.plot(np.deg2rad(np.array([180, 160, 140, 120])),
        [3, 3, 2.5, 2],'k', lw=4, marker='o', ms=10, zorder=3)
ax.set_ylabel('Geschwindigkeit [kn]')
ax.yaxis.set_label_coords(.98, .7)

fig.savefig('../fig_out/Polar_erklaerung.pdf', bbox_inches='tight')
fig.savefig('../fig_out/Polar_erklaerung.png', bbox_inches='tight', resolution=300)
```

Die Geschwindigkeit, die ein Segelschiff erreichen kann, hangt vom Winkel ab, mit dem der Wind auf das das Schiff trifft. Inn der Seemanssprache gibt es dafur spezielle Ausdrucke. Wenn der Wind von vorn kommt (auf den Bug trifft), dann nennt man das "im Wind". Ein rahgetakeltes Schiff kann im Wind keinene Vortrieb machen. Trifft der Wind dagegen von hinten auf das Schifft, spricht man davon, dass das Schiff "vor dem Wind segelt". In dieser Position werden fur rahgetakelte Schiffe die hochsten Geschwindigkeiten erreicht. Die Abhangigkeit der Geschwindigkeit vom Windwinkel lasst sich gut ein einem Polardiagram darstellen. Dazu stellt man sich das Schiff in der Mitte eines Kreises vor. Dann werden alle Messwerte, die bei einer bestimmten Windgeschwindigkeit aufgezeichnet wurden (z.B. Wind zwischen 5 und 6 Knoten) als jeweils ein farbiger Punkt in diesem Diagram eingetragen. Der Abstand von der Mitte des Diagramms zeigt die Bootsgeschwindigkeit an, der Winkel gibt die Windrichtung an. In der Praxis streuen die Messwerte etwas, weil der Wind nicht immer gleichmassig ist, das Segel nicht immer ideal ausgerichtet ist etc. Deshalb fassen wir die Messwerte nach Sektoren zusammen und bilden einen Mittelwert. In der Grafik wird der Mittelwert der Geschwindigkeiten aller grunen, roten, blauen und gelben Punkte gebildet und mit einem schwarzen Punkt markiert. Um den Zusammehang zwischen Windwinkel und Bootsgeschwindigkeit besser mit dme Auge verfolgen zu konnen, werden die Punkte mit einer Linie verbunden. Bei unseren Messfahrren haben wir manchmal tausende Messpunkte. Deshalb werden in den Polardigrammen nur die gemittelten Linien eingezeichnet. Oft zeigen wir Linien fur verschiedne Windgeschwindigkeiten ein einem Diagram oder vergleichen verschiedne Schiffe oder verschiedene Segeltypen.
